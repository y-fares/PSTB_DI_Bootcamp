**What is data analysis?**

Data analysis means examining, cleaning, and interpreting data to find patterns and insights that guide decisions. It turns raw information into usable knowledge.

**Why it’s important today ?**

In a world overflowing with data, analysis helps companies and institutions make faster, smarter, and evidence-based choices. Without it, decisions rely on intuition instead of facts.

Three key applications

*   Healthcare: Predicting diseases, improving diagnostics, and optimizing hospital management.

Colonnes intéressantes à analyser selon le type d’étude

| Type d’analyse                                | Colonnes pertinentes               | Justification                                                                                                                                                                                   |
| --------------------------------------------- | ---------------------------------- | ----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Trend Analysis (Analyse de tendance)**      | `Year`, `Avg hrs per day sleeping` | Permet d’observer l’évolution du temps de sommeil moyen au fil des années. On peut visualiser cela avec une courbe temporelle pour identifier si les Américains dorment plus ou moins qu’avant. |
| **Group Comparison (Comparaison de groupes)** | `Sex`, `Age Group`, `Type of Days` | Idéal pour comparer les habitudes de sommeil selon le genre, la tranche d’âge ou les jours (travail, week-end, tous jours).                                                                     |
| **Reliability / Statistical Confidence**      | `Standard Error`                   | Donne une idée de la précision des estimations : plus l’erreur standard est faible, plus la moyenne est fiable.                                                                                 |
| **Contextual Filtering**                      | `Activity`, `Period`               | Permet de sélectionner uniquement les données liées au sommeil ou d’autres activités, sur une base annuelle ou journalière.                                                                     |
