```python
df_twitter = pd.read_csv("Twitter Tweets Sentiment Dataset/Tweets.csv")
```

```python
df_twitter.head()
```

```python
df_twitter.dtypes
```

```python
import re
```

```python
df_twitter['hashtags'] = df_twitter['text'].apply(lambda x: re.findall(r"#\w+", str(x)))
df_twitter['mentions'] = df_twitter['text'].apply(lambda x: re.findall(r"@\w+", str(x)))
df_twitter['text_length'] = df_twitter['text'].apply(lambda x: len(str(x)))
```

```python
structured_tweets = df_twitter[['textID', 'text', 'hashtags', 'mentions', 'text_length', 'sentiment']]
```
